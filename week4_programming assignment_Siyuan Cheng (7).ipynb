```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import ipywidgets
```

```python
# set column name
ufo_country = pd.read_csv("https://github.com/UIUC-iSchool-DataViz/is445_AOUAOG_fall2021/raw/master/week05/data/ufo-scrubbed-geocoded-time-standardized-00.csv",
                  names = ["date", "city", "state", "country",
                          "shape", "duration_seconds", "duration",
                          "comment", "report_date", "latitude", "longitude"],
                  parse_dates = ["date", "report_date"])
```

```python
ufo_country
```

```python
# correct NAN
ufo_country = ufo_country.fillna({'country':'unknown', 'state': 'unknown', 'shape':'unknown'})
```

```python
ufo_country
```

```python
# set sampling
import random

# set a seed
random.seed(1234)
# 5000 samples
nsamples = 5000
small_sample = np.random.choice(range(len(ufo_country)-1), nsamples, replace=False)
```

```python
small_sample
```

```python
ufo_small = ufo_country.loc[small_sample]
```

```python
ufo_small
```

```python
# count the variable "country"
sightings = ufo_small.groupby(['country'])['city'].describe()
sightings
```

```python
# sort the data 
ufo_count = sightings["count"]
ufo_order = ufo_count.sort_values()
ufo_order
```

```python
@ipywidgets.interact(style = plt.style.available)
def make_plot(style):
    with plt.style.context(style):
        ax = ufo_order.plot(kind = "bar", rot=90, label='')
        ax.set_xlabel("countries")
        ax.set_ylabel("number of UFO sightings")
    plt.show()
 
```

# Explanation

I would like choose bar chart here. The reason is I think only bar chart can clearly expresses the difference. If I change to
